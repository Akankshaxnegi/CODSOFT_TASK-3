```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# LOADING  THE  DATASET
credit = pd.read_csv('/content/creditcard.csv')
```

```python
# Show the first five columns of the dataset " creditcard.csv"
credit.head()
```

```python
# Show the last five columns of the dataset " creditcard.csv"
credit.tail()
```

```python
# Displaying the statistical data of the dataset
credit.describe()
```

```python
# Viewing some more information to handle missing value (if any)
credit.info()
```

```python
# Checking the missing values
credit.isnull().sum()
```

```python
# Handling the missing values
credit_fraud= credit.fillna(value = 0)
credit_fraud
```

```python
# Checking again for missing values (if  any)
credit_fraud.isnull().sum()
```

```python
# Counting the distribution of normal transaction & fraud transaction (Normal = 9927 ,  Fraud = 38)
credit_fraud['Class'].value_counts()
```

```python
# Separating the transactions for further analysis
normal = credit_fraud[credit_fraud['Class'] == 0.0]
fraud = credit_fraud[credit_fraud['Class'] == 1.0]
print(normal.shape)
print(fraud.shape)
```

```python
# Statistical Measure of the normal transaction
normal.Amount.describe()
```

```python
# Statistical Measure of the fraud transaction
fraud.Amount.describe()
```

```python
# Comparing both transactions
credit_fraud.groupby('Class').mean()
```

**BUILD  A  SAMPLE  DATASET**

Creating a sample dataset with similar distribution like that in NORMAL & FRAUD Transactions.

No.  of  fraud  transaction = 38

```python
normal_sample = normal.sample(n=38)
```

```python
# Concatenating two dataframes
data = pd.concat([normal_sample , fraud], axis=0)
data.head()
```

```python
# Counting the distribution of normal_sample transaction & fraud transaction (Normal = 38 ,  Fraud = 38)
data['Class'].value_counts()
```

```python
# Comparing both transactions
data.groupby('Class').mean()
```

**SPLITTING  THE  DATA**

```python
x= data.drop(columns='Class' , axis=1)
y=data['Class']
print(x)
```

```python
print(y)
```

```python
# Splitting the data for TRAINING & TESTING
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , stratify= y , random_state=2)
print(x.shape , x_train.shape , x_test.shape)
```

**MODEL  TRAINING**

```python
# Training the LogisticRegression model with data
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train , y_train)
```

```python
# Accuracy on trained data
from sklearn.metrics import accuracy_score
x_train_predict = model.predict(x_train)
accuracy = accuracy_score(x_train_predict , y_train)
print("ACCURACY  ON  TRAINED DATA : ",accuracy)
```

```python
# Accuracy on test data
x_test_predict = model.predict(x_test)
accuracy = accuracy_score(x_test_predict , y_test)
print("ACCURACY ON TEST DATA : ",accuracy)
```
